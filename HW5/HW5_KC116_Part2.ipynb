```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q='

import urllib2
import urllib
import StringIO
import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()

def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))
```

```python
# average trip duration during weekends

query= '''
SELECT start_station_id,
starttime,
AVG(tripduration) AS trip_avg
FROM htv210.citibike_feb_2015
WHERE EXTRACT(DOW FROM starttime) IN (0,6)
GROUP BY start_station_id, starttime, usertype, gender
LIMIT 10'''

table = get_data(query)
table.head(5)
```

```python
#â€¢ Can we do the same for weekday?

query= '''
SELECT start_station_id,
starttime,
AVG(tripduration) AS trip_avg
FROM htv210.citibike_feb_2015
WHERE EXTRACT(DOW FROM starttime) NOT IN (0,6)
GROUP BY start_station_id, starttime, usertype, gender
LIMIT 10'''

table = get_data(query)
table.head(5)
```

```python
# Showing the list of start station locations 

query= '''
SELECT CDB_TransformToWebmercator(CDB_LatLng(start_station_latitude,start_station_longitude)) as the_geom_webmercator,
MIN(cartodb_id) as cartodb_id,
MIN(start_station_name) as start_station_name FROM htv210.citibike_feb_2015
GROUP BY start_station_latitude,start_station_longitude
'''
table = get_data(query)
table.head(5)
```

```python
#â€¢ Showing the number of trips started per station 
query = '''
SELECT COUNT(tripduration) AS nums,
CDB_TransformToWebmercator(
       CDB_LatLng(start_station_latitude,start_station_longitude)) as the_geom_webmercator,
       MIN(cartodb_id) as cartodb_id,
       MIN(start_station_name) as start_station_name FROM htv210.citibike_feb_2015
GROUP BY start_station_latitude,start_station_longitude
'''
table = get_data(query)
table.head(5)
```

```python
# Showing the number of trips started per station
# but only for stations within 500m of Time Square
# The coordinates of Time Square is (40.7577,-73.9857)

query = '''
SELECT start_station_id,
CDB_TRANSFORMTOWEBMERCATOR(CDB_LatLng(
start_station_latitude,
start_station_longitude
))
AS THE_GEOM_WEBMERCATOR,
MIN(cartodb_id) AS carto_id,
COUNT(tripduration) AS trip_count
FROM htv210.citibike_feb_2015
WHERE ST_DWITHIN(CDB_LatLng(
start_station_latitude,
start_station_longitude)::GEOGRAPHY,
CDB_LatLng(40.7577, -73.9857)::GEOGRAPHY, 500
)

GROUP BY start_station_id, start_station_latitude, start_station_longitude
LIMIT 10'''

table = get_data(query)
table.head(5)
```
