```python
#read the csv
df = pd.read_csv(os.getenv("PUIDATA")+'/'+'death.csv')

df.head()
```

```python
#df['Age Adjusted Death Rate'].value_counts()
```

```python
# This drops all empty/NaN values
df2 = df.loc[(df.Deaths != '.')].copy()
```

```python
#converts number of deaths to an int (previously an obj)
df2.Deaths = df2.Deaths.astype('int64')

#Total sum of deaths in dataset
df2.Deaths.sum()
```

```python
# This groups the number of deaths per year
df2.groupby(['Year'])['Deaths'].sum().plot()
```

```python
#Using Seaborn, this graph shows cause of death and the number of deceased per cause (not the final plot)

import seaborn as sns

sns.countplot(y='Leading Cause', data=df)
```

```python
# Examines leading cause of death for M/F
df3 = pd.DataFrame(df2.groupby(['Leading Cause', 'Sex'])['Deaths'].count())

df3 = df3.reset_index()

df3
```

Figure 1: The plot shows the number of deaths classified as 'accidental' for both male and females in NYC. The dataframe came from NYC OpenData. To make the plot, all columns were dropped except 'Accidents Except Drug Posioning (V01-X39, X43, X45-X59, Y85-Y86)' and 'Mental and Behavioral Disorders due to Accidental Poisoning and Other Psychoactive Substance Use (F11-F16, F18-F19, X40-X42, X44)'. The legends were shortened to better serve the aesthetics of the plot. Overall, the plot shows a higher number of accidental deaths for men than women. The data does not include non-binary gender identifications.
